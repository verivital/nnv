```python
# Configuration - Docker container paths
BASE_DIR = "/nnv/code/nnv/examples/Submission/FORMALISE2025"
DATA_DIR = f"{BASE_DIR}/data"
MODELS_DIR = f"{BASE_DIR}/models"
```

```python
import numpy as np
import os
```

```python
NUM_FRAMES = 32
```

```python
data_dir = DATA_DIR
```

```python
train_data = np.load(os.path.join(data_dir, 'KTHActions', str(NUM_FRAMES), f'kthactions_training_{NUM_FRAMES}.npy'))
val_data = np.load(os.path.join(data_dir, 'KTHActions', str(NUM_FRAMES), f'kthactions_val_{NUM_FRAMES}.npy'))
test_data = np.load(os.path.join(data_dir, 'KTHActions', str(NUM_FRAMES), f'kthactions_test_{NUM_FRAMES}.npy'))
```

```python
test_data.shape
```

```python
def convert_to_grayscale(videos):
  R = videos[:, :, :, :, 0]
  G = videos[:, :, :, :, 1]
  B = videos[:, :, :, :, 2]

  grayscale_videos = 0.2989 * R + 0.5870 * G + 0.1140 * B

  return grayscale_videos
```

```python
grayscale_train_data = convert_to_grayscale(train_data)
grayscale_val_data = convert_to_grayscale(val_data)
grayscale_test_data = convert_to_grayscale(test_data)
```

# Normalize the data before saving

```python
normalized_grayscale_train_data = grayscale_train_data / 255.0
normalized_grayscale_val_data = grayscale_val_data / 255.0
normalized_grayscale_test_data = grayscale_test_data / 255.0
```

```python
grayscale_save_dir = os.path.join(DATA_DIR, 'KTHActions', 'grayscale', str(NUM_FRAMES))
os.makedirs(grayscale_save_dir, exist_ok=True)

np.save(os.path.join(grayscale_save_dir, f'kthactions_grayscale_{NUM_FRAMES}f_train_data.npy'), normalized_grayscale_train_data)
np.save(os.path.join(grayscale_save_dir, f'kthactions_grayscale_{NUM_FRAMES}f_val_data.npy'), normalized_grayscale_val_data)
np.save(os.path.join(grayscale_save_dir, f'kthactions_grayscale_{NUM_FRAMES}f_test_data.npy'), normalized_grayscale_test_data)
```
